```python
import pandas as pd
```

```python
df1 = pd.read_excel('data.xlsx')

#Data sheet#
```

```python
df2 = df1[['Order_ID', 'SERVER','TOTAL']]
#Order_ID, Server and Total bill amount for Order_Id are required to find efficient server#
```

```python
import numpy as np
```

```python
group = df2.groupby('Order_ID')
df3 = group['TOTAL'].agg([np.sum])
#Order_ID and total bill amount for that order#
df3
```

```python
df4 = df2[['Order_ID','SERVER']]
df4
```

```python
df4 = df4.drop_duplicates()
df4
#Order_ID and server who took the order#
```

```python
group2 = df2.groupby('SERVER')
df5 = group2['TOTAL'].agg([np.sum])
```

```python
df5['SERVER_NAME'] = df5.index
df5
#Server and total bill revenue generated by the server#
```

```python
df6 = df4['SERVER'].value_counts()
df6.columns = ['server_name','num_of_bills']

```

```python
df6 = df6.to_frame()
df6['SERVER_NAME']=df6.index
df6
#Server name and number of bills generated by the server#
```

```python
df7 = pd.merge(df5, df6, on='SERVER_NAME', how='outer')
```

```python
df7['avg_bill'] = df7['sum']/df7['SERVER']
```

```python
df7 = df7.rename(columns={"sum": "Total revenue generated", "SERVER_NAME": "Server Name", "SERVER" : "Customers served", "avg_bill" : "Average revenue per bill"})
df7
```

```python
df8 = df7.sort_values(by='Average revenue per bill', ascending=False)
df8
#Average revenue generated per bill by each server#
```

```python
b = df8['Customers served'][8:].sum()
b
```
